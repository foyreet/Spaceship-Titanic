# Обучение модели

```python
X = df.drop(columns=['Transported'])
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

# --- Определяем базовые модели ---
base_models = [
    ('rf', RandomForestClassifier(n_estimators=50, random_state=42)),
    ('gb', GradientBoostingClassifier(n_estimators=50, random_state=42))
]

# --- Определяем мета-модель ---
meta_model = LogisticRegression()

# --- Стек-модель с k-fold ---
stacking_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    cv=KFold(n_splits=3, shuffle=True, random_state=42)  # здесь k-fold
)

# --- Оценка модели через k-fold ---
scores = cross_val_score(stacking_model, X, y, cv=KFold(n_splits=3, shuffle=True, random_state=42), scoring='accuracy')
print("Средняя точность:", np.mean(scores))
```

```python
# --- Финальное обучение ---
stacking_model.fit(X, y)

# --- Предсказание ---
y_test_pred = stacking_model.predict(test)
```

```python
sample_submission = pd.read_csv("/kaggle/input/spaceship-titanic/sample_submission.csv")
```

```python
submission = pd.DataFrame({
    "PassengerId": sample_submission["PassengerId"],
    "Transported": y_test_pred.astype(bool)
})
submission.to_csv("submission.csv", index=False)
```
